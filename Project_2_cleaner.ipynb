```python
import pandas as pd
```

```python
csvinput = "./WeatherData.csv"
csvoutput = "./WeatherData_cleaned.csv"
```

```python
df = pd.read_csv (csvinput)
```

```python
df.head()
```

```python
df.shape
```

```python
def dateparse(date):
    datelist = int(date.split("-")[0])
    return datelist
df["year"] = df["StartTime(UTC)"].map(dateparse)
```

```python
df_2017 = df.loc[df["year"]==2017,:]
```

```python
df_2017
```

```python
df_2017.to_csv(csvoutput,index=False)
```
