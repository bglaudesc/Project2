```python
import pandas as pd
```

```python
csvinput = "./WeatherData.csv"
csvoutput = "./WeatherData_cleaned.csv"
```

```python
df = pd.read_csv (csvinput)
```

```python
df.head()
```

```python
df.shape
```

```python
def year(date):
    datelist = int(date.split("-")[0])
    return datelist
df["year"] = df["StartTime(UTC)"].map(year)
```

```python
def month(date):
    datelist = int(date.split("-")[1])
    return datelist
df["month"] = df["StartTime(UTC)"].map(month)
```

```python
df_2017 = df.loc[df["year"]==2017,:].drop(columns=[
    "EventId",
    "StartTime(UTC)", 
    "EndTime(UTC)",
    "TimeZone", 
    "year"
])
```

```python
def Severity(label):
    if label=="Light" or label=="UNK":
        return 1
    elif label=="Moderate":
        return 2
    elif label=="Heavy":
        return 3
    elif label=="Severe" or label=="Other":
        return 4 
    
df_2017["Severity"] = df_2017["Severity"].map(Severity)
```

```python
# df_Type = pd.get_dummies(df_2017, columns=["Type"], prefix=[""], prefix_sep="" )
# df_Type

df_Severity_key = pd.DataFrame({
    "Severity_label": ["Light" , "UNK" , "Moderate" , "Heavy" , "Severe" , "Other"],
    "Severity_score": [1, 1, 2, 3, 4, 4]
})
df_Severity_key
```

```python
df_2017
```

```python
df_location = df_2017.drop(columns=["Type" , "Severity"]).drop_duplicates()
df_location
```

```python
Groupby_2017 = df_2017.groupby(["month", "AirportCode", "Type"])
EventType = Groupby_2017["Severity"].count()
EventType
```

```python
Groupby_Type = df_2017.groupby(["month", "AirportCode"])
EventTotal = Groupby_Type["Severity"].count()
EventTotal
```

```python
Avg_Severity = Groupby_Type["Severity"].mean()
```

```python
df_One = pd.DataFrame(EventType).reset_index().rename(columns={"Severity": "Type_Count"})
df_One = pd.pivot_table(df_One, index=["month", "AirportCode"],columns=["Type"] , values="Type_Count")\
    .fillna(0).reset_index()
df_One.columns.name=None
df_One

df_Two = pd.DataFrame({
    "EventTotal": EventTotal,
    "Avg_Severity": Avg_Severity
}).reset_index()
# df_Two
df_Summary = df_One.merge(df_Two , how="inner", on=["month", "AirportCode"])
df_Summary = df_Summary.merge(df_location , how="inner", on=["month", "AirportCode"]).drop(columns=["AirportCode"])
df_Summary
```

```python
df_2017.to_csv(csvoutput,index=False)
df_Summary.to_csv("final_weather.csv",index=False)
```
