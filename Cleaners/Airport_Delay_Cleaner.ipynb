```python
import pandas as pd
```

```python
csvinput = "./Airline_Delay.csv"
csvoutput = "./Airline_Delay_cleaned.csv"
```

```python
df = pd.read_csv (csvinput)
```

```python
df.head()
```

```python
Delay_Groupby = df.groupby(["month" , "City" , "Air_code", "State", "Airport"]).sum()
Delay_Groupby
```

```python
df_Delay = pd.DataFrame(Delay_Groupby).reset_index()
df_Delay
```

```python
df_Delay.to_csv(csvoutput,index=False)
```
